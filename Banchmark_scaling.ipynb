```python
# ============================================================
# CELL 1 — imports & base config
# ============================================================
import time
import numpy as np
import matplotlib.pyplot as plt
from src.simulation import simulate, compute_exp_checkpoints

BASE_INFO = {
    "General": {"seed": 42},
    "Graph": {"type": "ER", "n": 200, "p": 0.05},
    "OD": {"model": "BCM", "epsilon": 0.3, "mu": 0.1},
    "Ranker": {"rule": "Random"},
    "Simulation_details": {"n_steps": 10000, "convergence_window": 50, "convergence_delta": 1e-4},
    "post_history": 50,
    "k_posts": 1,
}

def make_info(**overrides):
    import copy
    info = copy.deepcopy(BASE_INFO)
    for k, v in overrides.items():
        if isinstance(v, dict):
            info[k].update(v)
        else:
            info[k] = v
    return info

def get_checkpoints(info):
    n_steps = info["Simulation_details"]["n_steps"]
    return set(compute_exp_checkpoints(n_steps))

def bench(info, n_runs=3):
    homophily_steps = get_checkpoints(info)
    times = []
    conv_steps = []
    for run in range(n_runs):
        t0 = time.perf_counter()
        result = simulate(info, seed=42+run, homophily_steps=homophily_steps)
        times.append(time.perf_counter() - t0)
        conv_steps.append(result['convergence_step'])
    return np.mean(times), np.std(times), np.mean(conv_steps), np.min(conv_steps), np.max(conv_steps)

def bench_with_opinions(info, n_runs=3):
    homophily_steps = get_checkpoints(info)
    times = []
    conv_steps = []
    all_opinions = []
    for run in range(n_runs):
        t0 = time.perf_counter()
        result = simulate(info, seed=42+run, homophily_steps=homophily_steps)
        times.append(time.perf_counter() - t0)
        conv_steps.append(result['convergence_step'])
        all_opinions.append(result['opinions'])  # shape: (n_steps, n_users)
    return np.mean(times), np.std(times), np.mean(conv_steps), np.min(conv_steps), np.max(conv_steps), all_opinions, result
```
